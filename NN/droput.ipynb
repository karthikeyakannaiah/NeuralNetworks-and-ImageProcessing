```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
 
df = pd.read_csv('./assets/sonar.csv',header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
X = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
from keras import Sequential, layers
```

```python
model = Sequential([
    layers.Dense(60, input_dim=60, activation='relu'),
    layers.Dense(30, activation='relu'),
    layers.Dense(15, activation='relu'),
    layers.Dense(1,activation='sigmoid')
])

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.fit(X_train,y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
model2 = Sequential([
    layers.Dense(60, input_dim=60, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(30, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(15, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1,activation='sigmoid')
])

model2.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model2.fit(X_train,y_train, epochs=100, batch_size=8)
model2.evaluate(X_test,y_test)
```

```python
y_pred2 = model2.predict(X_test).reshape(-1)
y_pred2 = np.round(y_pred2)
print(classification_report(y_test, y_pred2))
```
